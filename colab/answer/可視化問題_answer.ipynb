## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
hist_data = pd.read_csv("https://raw.githubusercontent.com/tikutikuwa/Basic-Python/master/colab/quizzes/matplotlib_quiz/hist_data.csv")
plt.hist(hist_data["aデータ"], bins=30, color="orange", label="aデータ", alpha=0.5)
plt.hist(hist_data["bデータ"], bins=30, color="turquoise", label="bデータ", alpha=0.5)
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
scatter_data = pd.read_csv("https://raw.githubusercontent.com/tikutikuwa/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv")
class0 = scatter_data.loc[scatter_data["class"]==0]
class1 = scatter_data.loc[scatter_data["class"]==1]
class2 = scatter_data.loc[scatter_data["class"]==2]
plt.scatter(class0["length (cm)"], class0["width (cm)"], c="r", marker="o", label="class0")
plt.scatter(class1["length (cm)"], class1["width (cm)"], c="b", marker="o", label="class1")
plt.scatter(class2["length (cm)"], class2["width (cm)"], c="g", marker="o", label="class2")
plt.title("クラス０～２のlengthとwidthに関する散布図")
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.grid(True)
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
line_data = pd.read_csv("https://raw.githubusercontent.com/tikutikuwa/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv")
fig = plt.figure(figsize=(6,10))
plt.subplots_adjust(hspace=0.5)
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(line_data["期間"], line_data["件数"], marker=".", ms=10)
ax1.tick_params(axis='x', labelrotation=30)
ax1.grid(True)
ax1.set_title("期間ごとの件数の推移")
ax1.set_xlabel("期間")
ax1.set_ylabel("件数")

ax2.bar(line_data["期間"], line_data["費用"])
ax2.tick_params(axis='x', labelrotation=30)
ax2.grid(True)
ax2.set_title("期間ごとの費用の推移")
ax2.set_xlabel("期間")
ax2.set_ylabel("費用(円)")
```
